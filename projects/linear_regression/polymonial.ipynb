```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
m = 100
X = 6 * np.random.rand(m, 1) -3
y = 0.5 * X**2 + X + 2 + np.random.randn(m, 1)
```

```python
plt.scatter(X, y, label='Points')
plt.legend()
plt.xlabel('Axis x')
plt.ylabel('Axis y')
plt.title('Data')
plt.show()
```

```python
poly_features = PolynomialFeatures(degree=3, include_bias=False)
X_poly = poly_features.fit_transform(X)
```

```python
linear_regresssor = LinearRegression()
linear_regresssor.fit(X_poly, y)
predictions = linear_regresssor.predict(X_poly)
```

```python
plt.scatter(X, y, label='Points')
plt.legend()
plt.xlabel('Axis x')
plt.ylabel('Axis y')
plt.title('Data')

X_new=np.linspace(-3, 3, 100).reshape(100, 1)
X_new_poly = poly_features.transform(X_new)
y_new = linear_regresssor.predict(X_new_poly)

plt.plot(X_new, y_new, "r-", linewidth=2, label="Predictions")
plt.show()
```

```python
training_model_rmse = np.sqrt(mean_squared_error(y, predictions))
training_model_rmse
```

```python
linear_regresssor.score(X_poly, y)
```
