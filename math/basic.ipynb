```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
from math import log
```

Drawing chart

```python

x = np.linspace(-10, 10, 100)

y = 2 * x + 1

plt.plot(x, y)

plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

plt.xlabel('x')
plt.ylabel('y')

plt.title('y = 2x + 1')

plt.show()
```

```python

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

X, Y = np.meshgrid(x, y)

Z = 2 * X + 3 * Y

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z)

ax.set_xlabel('O x')
ax.set_ylabel('O y')
ax.set_zlabel('O z')

plt.title('f(x, y) = 2x + 3y')

plt.show()
```

Derivative

```python
x = sp.Symbol('x')
f = x**2 + 1

df = sp.diff(f, x)

print(df)
```

Derivative in a point

```python
x = sp.Symbol('x')
a = 2

f = x**2

df = sp.diff(f, x).subs(x, a)

print(df)
```

Partial derivative

```python
from sympy.plotting import plot3d

x,y = sp.symbols('x y')
f = 2*x**3 + 3*y**3
dx_f = sp.diff(f, x)
dy_f = sp.diff(f, y)

print(dx_f, dy_f)

plot3d(f)
```

Chain deriverate

```python
x, y = sp.symbols('x y')
y = x**2 + 1
c = x**3 - 2
z = y**3 - 2
v = (x**2 + 1)**3 - 2

# dz/dx = y * c (deriverate)

df_x = sp.diff(y, x)
df_c = sp.diff(c, x)
df_z = sp.diff(z, x)
df_chain = sp.diff(v, x)
print(df_x)
print(df_c)
print(df_z)
print(df_chain)
```

Calculate integral

```python
x = sp.symbols('x')
f = x**2 + 5

field = sp.integrate(f, (x, 0, 1))
field
```
