```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

x = np.linspace(-10, 10, 100)

y = 2 * x + 1

plt.plot(x, y)

plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

plt.xlabel('x')
plt.ylabel('y')

plt.title('y = 2x + 1')

plt.show()
```

```python

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

X, Y = np.meshgrid(x, y)

Z = 2 * X + 3 * Y

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z)

ax.set_xlabel('O x')
ax.set_ylabel('O y')
ax.set_zlabel('O z')

plt.title('f(x, y) = 2x + 3y')

plt.show()
```
